```python
import requests
import datetime

valcode = 'USD'

end = datetime.datetime.today()
start = end - datetime.timedelta(days=365)

rates_by_month = {}

for i in range((end - start).days):
    date = start + datetime.timedelta(days=i)
    month = date.month
    year = date.year
    params = {
        'date': date.strftime("%Y%m%d"),
        'valcode': valcode,
        'json': "json"
    }

    r = requests.get(f'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange', params=params)
    data = r.json()
    rate = data[0]['rate']

    if (year, month) not in rates_by_month:
        rates_by_month[(year, month)] = [rate]
    else:
        rates_by_month[(year, month)].append(rate)

for year_month, rates in rates_by_month.items():
    year, month = year_month
    avg_rate = sum(rates) / len(rates)
    print(f'Середній курс {valcode} за {month}/{year}: {avg_rate}')

deviations = []
for year_month, rates in rates_by_month.items():
    year, month = year_month
    avg_rate = sum(rates) / len(rates)
    for rate in rates:
        deviation = rate - avg_rate
        deviations.append((year, month, rate, deviation))

for deviation in deviations:
    year, month, rate, deviation_value = deviation
    print(f'{valcode} курс на {month}/{year}: {rate}, відхилення від середнього: {deviation_value}')
```

```python
import pickle

# Збереження списку в файл
with open('deviations.pkl', 'wb') as f:
    pickle.dump(deviations, f)

# Завантаження списку з файлу
with open('deviations.pkl', 'rb') as f:
    deviations = pickle.load(f)
```
