```python
import random

##функція для мин и макс
def find_min_max(L):

    max = L[0]
    min = L
    for i in L:
        if i<min:
            min=i
        if i>max:
            max=i
    return (min, max)


import math
from statistics import median

##функція для середнього
def mean(L):
    return sum(L) / len(L)

##функція для середньоквадратичного
def standard_deviation(L):
    n = len(L)
    if n < 2:
        return 0
    mean_value = sum(L) / n
    variance = sum((x - mean_value) ** 2 for x in L) / (n - 1)
    return math.sqrt(variance)

##функція для медіани
def median(L):
    return median(L)
```
